```python
import re,collections
```

```python
#helper functions
def pair_count_gen(word_dict):   #generates the dictionary of hyphen sep words and their corresponding count
  pair = collections.defaultdict(int)
  for word,count in word_dict.items():
    characters = word.split("-")
    for i in range(len(characters)-1):
      pair[characters[i],characters[i+1]] += count
  return pair


def merge(freq_char, word_dict):      #source: research paper from the assignment
    new_word_dict = {}
    bigram = '-'.join(freq_char)
    p = re.compile(r'(?:(?<=^)|(?<=-))' + re.escape(bigram) + r'(?:(?=-|$))')
    for word in word_dict:
        w_out = p.sub(''.join(freq_char), word)
        new_word_dict[w_out] = word_dict[word]
    return new_word_dict


def word_hyphen(s):    #takes string and return s-t-r-i-n-g-$
  if(len(s)==1):
    return s
  out = ""
  for i in s:
    out = out + (i+"-")
    # print(out)
  out = out[0:len(out)-1]
  out+="-$"
  return out

```
